```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Fetch revenue data
symbol = 'EL'
EsteeLauder = yf.Ticker(symbol)

# Get income statement data (focus on revenues)
income_statement = EsteeLauder.financials
revenue_data = income_statement.loc['Total Revenue'].transpose()

# Display the revenue data
print("Estee Lauders's Revenue Data (in billions):")
print(revenue_data)

# Check for NaN values in revenue data
if revenue_data.isnull().any():
    print("\nWarning: NaN values detected in revenue data. Handling them...")
    revenue_data = revenue_data.dropna()  # Drop rows with NaN values
    # Alternatively, you could fill NaN values with the mean or median:
    # revenue_data = revenue_data.fillna(revenue_data.mean())

# Prepare data for regression
# Convert index to datetime for better handling
revenue_data.index = pd.to_datetime(revenue_data.index)
revenue_data = revenue_data.sort_index()

# Create a numerical index for regression
X = np.array(range(len(revenue_data))).reshape(-1, 1)  # Independent variable (time)
y = revenue_data.values  # Dependent variable (revenue)

# Fit a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict future revenues (for the next 3 years)
future_X = np.array(range(len(revenue_data), len(revenue_data) + 3)).reshape(-1, 1)  # Next 3 years
predicted_revenue = model.predict(future_X)

# Combine historical and predicted revenues
future_dates = pd.date_range(start=revenue_data.index[-1] + pd.DateOffset(years=1), periods=3, freq='Y')
predicted_revenue_series = pd.Series(predicted_revenue, index=future_dates)

# Plot historical and predicted revenues
plt.figure(figsize=(10, 5))
plt.plot(revenue_data.index, revenue_data.values, label='Historical Revenue', marker='o')
plt.plot(predicted_revenue_series.index, predicted_revenue_series.values, label='Predicted Revenue', marker='o', linestyle='--')
plt.title("Estee Lauders's Revenue Prediction")
plt.xlabel("Year")
plt.ylabel("Revenue (in billions)")
plt.legend()
plt.grid()
plt.show()

# Print predicted revenues
print("\nPredicted Revenues for the Next 3 Years:")
for year, revenue in zip(predicted_revenue_series.index.year, predicted_revenue_series.values):
    print(f"{year}: ${revenue:.2f} billion")
```

```python
# Look at Financial Ratios

import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd  # Import pandas

# Define the ticker symbol for Apple Inc.
ticker_symbol = 'EL'

# Fetch the data
EsteeLauder = yf.Ticker(ticker_symbol)

# Get the balance sheet and income statement
balance_sheet = EsteeLauder.balance_sheet.transpose()
income_statement = EsteeLauder.financials.transpose()

# Convert the index to datetime for easier plotting
balance_sheet.index = pd.to_datetime(balance_sheet.index)
income_statement.index = pd.to_datetime(income_statement.index)

# Select important items for calculations
total_assets = balance_sheet['Total Assets']
total_liabilities = balance_sheet['Total Liabilities Net Minority Interest']
shareholder_equity = total_assets - total_liabilities  # Calculate Shareholder's Equity

net_income = income_statement['Net Income']
ordinary_shares = balance_sheet['Ordinary Shares Number']

# Calculate metrics
leverage_ratio = total_liabilities / total_assets
roe = net_income / shareholder_equity
eps = net_income / ordinary_shares

# Create a DataFrame for metrics
metrics = pd.DataFrame({
    'Leverage Ratio': leverage_ratio,
    'Return on Equity (ROE)': roe,
    'Earnings Per Share (EPS)': eps
})

# Plotting Leverage Ratio
plt.figure(figsize=(12, 6))
plt.plot(metrics.index, metrics['Leverage Ratio'], marker='o', color='blue')
plt.title(f'Leverage Ratio Over Time for {ticker_symbol}')
plt.xlabel('Date')
plt.ylabel('Leverage Ratio')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()

# Plotting Return on Equity (ROE)
plt.figure(figsize=(12, 6))
plt.plot(metrics.index, metrics['Return on Equity (ROE)'], marker='o', color='green')
plt.title(f'Return on Equity (ROE) Over Time for {ticker_symbol}')
plt.xlabel('Date')
plt.ylabel('Return on Equity (ROE)')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()

# Plotting Earnings Per Share (EPS)
plt.figure(figsize=(12, 6))
plt.plot(metrics.index, metrics['Earnings Per Share (EPS)'], marker='o', color='orange')
plt.title(f'Earnings Per Share (EPS) Over Time for {ticker_symbol}')
plt.xlabel('Date')
plt.ylabel('Earnings Per Share (EPS)')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Fetch income statement data for Estee Lauder
symbol = 'EL'
estee_lauder = yf.Ticker(symbol)
income_statement = estee_lauder.financials

```

```python


# Extract relevant data
revenue_data = income_statement.loc['Total Revenue'].transpose()
cogs_data = income_statement.loc['Cost Of Revenue'].transpose()
net_income_data = income_statement.loc['Net Income'].transpose()

# Check for NaN values and handle them
for data in [revenue_data, cogs_data, net_income_data]:
    if data.isnull().any():
        data.dropna(inplace=True)

```

```python
# Prepare data for regression
def prepare_data(data):
    data.index = pd.to_datetime(data.index)
    data = data.sort_index()
    X = np.array(range(len(data))).reshape(-1, 1)
    y = data.values
    return X, y

X_revenue, y_revenue = prepare_data(revenue_data)
X_cogs, y_cogs = prepare_data(cogs_data)
X_net_income, y_net_income = prepare_data(net_income_data)

# Fit linear regression models
model_revenue = LinearRegression().fit(X_revenue, y_revenue)
model_cogs = LinearRegression().fit(X_cogs, y_cogs)
model_net_income = LinearRegression().fit(X_net_income, y_net_income)

```

```python
# Predict future values for the next 3 years
future_X = np.array(range(len(revenue_data), len(revenue_data) + 3)).reshape(-1, 1)
predicted_revenue = model_revenue.predict(future_X)
predicted_cogs = model_cogs.predict(future_X)
predicted_net_income = model_net_income.predict(future_X)

```

```python
# Combine historical and predicted data
future_dates = pd.date_range(start=revenue_data.index[-1] + pd.DateOffset(years=1), periods=3, freq='Y')
predicted_revenue_series = pd.Series(predicted_revenue.flatten(), index=future_dates)
predicted_cogs_series = pd.Series(predicted_cogs.flatten(), index=future_dates)
predicted_net_income_series = pd.Series(predicted_net_income.flatten(), index=future_dates)

# Plot historical and predicted data
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(revenue_data.index, revenue_data.values, label='Historical Revenue', marker='o')
plt.plot(predicted_revenue_series.index, predicted_revenue_series.values, label='Predicted Revenue', marker='o', linestyle='--')
plt.title("Estee Lauder's Revenue Prediction")
plt.xlabel("Year")
plt.ylabel("Revenue (in billions)")
plt.legend()
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(cogs_data.index, cogs_data.values, label='Historical COGS', marker='o')
plt.plot(predicted_cogs_series.index, predicted_cogs_series.values, label='Predicted COGS', marker='o', linestyle='--')
plt.title("Estee Lauder's COGS Prediction")
plt.xlabel("Year")
plt.ylabel("COGS (in billions)")
plt.legend()
plt.grid()

plt.subplot(1, 3, 3)
plt.plot(net_income_data.index, net_income_data.values, label='Historical Net Income', marker='o')
plt.plot(predicted_net_income_series.index, predicted_net_income_series.values, label='Predicted Net Income', marker='o', linestyle='--')
plt.title("Estee Lauder's Net Income Prediction")
plt.xlabel("Year")
plt.ylabel("Net Income (in billions)")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Step 1: Download Estée Lauder Financials
el = yf.Ticker("EL")
```

```python
# -------------------------------
# Income Statement Forecast
# -------------------------------

# Extract Income Statement (annual)
income_stmt = el.financials.T  # Transposed to view by year
target_cols = ["Total Revenue", "Operating Income", "Net Income"]
selected_data = income_stmt[target_cols].dropna()

# Prepare data for modeling
years = pd.to_datetime(selected_data.index).year.values.reshape(-1, 1)
forecast_years = np.array([[2025], [2026], [2027]])
```

```python
# Plot and forecast each item
forecast_results = {}
plt.figure(figsize=(14, 6))

for i, col in enumerate(target_cols):
    y = selected_data[col].values
    model = LinearRegression().fit(years, y)
    forecast = model.predict(forecast_years)
    forecast_results[col] = forecast

    # Plot
    plt.subplot(1, 3, i + 1)
    plt.plot(years.flatten(), y, marker='o', label='Actual')
    plt.plot(forecast_years.flatten(), forecast, marker='x', linestyle='--', label='Forecast')
    plt.title(col)
    plt.xlabel("Year")
    plt.ylabel("USD")
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Forecast Summary
forecast_df = pd.DataFrame(forecast_results, index=[2025, 2026, 2027])
print("Forecasted Income Statement Values for Estée Lauder (in USD):")
print(forecast_df.applymap(lambda x: f"${x:,.0f}"))
```
